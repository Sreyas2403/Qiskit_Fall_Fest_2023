```python
import csv

def adaptive_sampling(input_file):
    # Read data from CSV file
    with open(input_file, 'r') as file:
        reader = csv.reader(file)
        columns = list(reader)

    # Extract temperature data from the 13th column ("TEMP") with adaptive sampling
    output_rows = []
    prev_temp = None
    threshold = 2

    for col in columns[1:]:  # Skip the header row
        try:
            current_temp = float(col[12])  # Index 12 corresponds to the 13th column ("TEMP")

            # Skip rows with "NA"
            if col[12] == "NA":
                continue

            # Check temperature change against the threshold
            if prev_temp is not None and abs(current_temp - prev_temp) >= threshold:
                output_rows.append((current_temp, abs(current_temp - prev_temp)))
                print(f"Temperature Change Detected! Data: {current_temp:.2f}, Change: {abs(current_temp - prev_temp):.2f}")

            prev_temp = current_temp

        except (ValueError, IndexError):
            print(f"Skipping invalid or missing value.")

    # Print the number of output values and efficiency
    num_output_values = len(output_rows)
    num_input_values = len(columns) - 1  # Assuming the first row is header
    efficiency = (num_output_values / num_input_values) * 100

    print(f"\nNumber of Output Values: {num_output_values}")
    print(f"Efficiency: {efficiency:.2f}%")

# Specify the input file outside the function
input_file = 'BeijingPM20100101_20151231.csv'

# Run the adaptive_sampling function with the specified input file
adaptive_sampling(input_file)
```
